# RAS Model Validation ? 2008?2009 Global Financial Crisis

The 2008?2009 financial crisis caused **real production drops** in fertilizer manufacturing worldwide ? factories shut down due to collapsing demand. This is a genuine **production shock**, making it a much better test case for the RAS model than trade-barrier shocks like sanctions.

**Key shock countries (2007 ? 2009):**
- Belarus K?O: ?52% (5,097 ? 2,435 kt)
- Russia P?O?: ?60% (2,807 ? 1,111 kt)
- Russia K?O: ?43% (7,277 ? 4,145 kt)
- Russia N: ?24% (7,203 ? 5,464 kt)
- Israel K?O: ?53% (2,056 ? 975 kt)

**Validation design:**
| Component | Source | Period |
|-----------|--------|--------|
| Baseline trade matrix T? | FAOSTAT bilateral exports | Avg. 2005?2007 |
| Pre-shock production & demand | FAOSTAT | **2007** (single year) |
| Post-shock production & demand | FAOSTAT | **2009** (single year) |
| Actual post-shock trade | FAOSTAT bilateral exports | **2009** |

## 1. Load Data & Build Baseline / Post-Shock Datasets

```python
# ??????????????????????????????????????????????????????????????????????????????
# Configuration
# ??????????????????????????????????????????????????????????????????????????????
T0_YEARS       = [2005, 2006, 2007]       # baseline trade matrix: avg 3 years
PRE_YEAR       = 2007                      # pre-shock production & demand
POST_YEAR      = 2009                      # post-shock (crisis year)
MIN_THRESHOLD  = 1000                      # tonnes

NUTRIENT_CODES = {'N': 3102, 'P': 3103, 'K': 3104}
NUTRIENT_NAMES = {'N': 'Nitrogen (N)', 'P': 'Phosphate (P?O?)', 'K': 'Potash (K?O)'}
ELEMENT_CODES  = {'Production': 5510, 'Import': 5610, 'Export': 5910, 'AgUse': 5157}

SHOCK_COUNTRIES = ['Russian Federation', 'Belarus', 'Ukraine', 'Israel',
                   'Germany', 'France', 'Canada']

# ??????????????????????????????????????????????????????????????????????????????
# Helpers
# ??????????????????????????????????????????????????????????????????????????????
def load_series_single(df, item_code, elem_code, year):
    """Load production or demand for a single year."""
    ycol = f'Y{year}'
    mask = (df['Item Code'] == item_code) & (df['Element Code'] == elem_code)
    sub = df.loc[mask, ['Area', ycol]].copy()
    sub[ycol] = pd.to_numeric(sub[ycol], errors='coerce')
    return sub.groupby('Area')[ycol].sum()

def load_series_avg(df, item_code, elem_code, years):
    """Load production or demand averaged over multiple years."""
    ycols = [f'Y{y}' for y in years]
    mask = (df['Item Code'] == item_code) & (df['Element Code'] == elem_code)
    sub = df.loc[mask, ['Area'] + ycols].copy()
    for c in ycols:
        sub[c] = pd.to_numeric(sub[c], errors='coerce')
    sub['avg'] = sub[ycols].mean(axis=1)
    return sub.groupby('Area')['avg'].sum()

def load_trade_matrix(df, item_code, elem_code, years):
    """Build bilateral trade matrix averaged over years."""
    ycols = [f'Y{y}' for y in years]
    ycols_avail = [c for c in ycols if c in df.columns]
    mask = (df['Item Code'] == item_code) & (df['Element Code'] == elem_code)
    sub = df.loc[mask, ['Reporter Countries', 'Partner Countries'] + ycols_avail].copy()
    for c in ycols_avail:
        sub[c] = pd.to_numeric(sub[c], errors='coerce')
    sub['avg'] = sub[ycols_avail].mean(axis=1)
    flows = sub[['Reporter Countries', 'Partner Countries', 'avg']].dropna(subset=['avg'])
    flows = flows[flows['avg'] > 0]
    return flows.pivot_table(
        index='Reporter Countries', columns='Partner Countries',
        values='avg', aggfunc='sum', fill_value=0
    )

def load_trade_single(df, item_code, elem_code, year):
    """Build bilateral trade matrix for a single year."""
    return load_trade_matrix(df, item_code, elem_code, [year])

# ??????????????????????????????????????????????????????????????????????????????
# Load CSVs
# ??????????????????????????????????????????????????????????????????????????????
print("Loading inputs data...")
df_inputs = pd.read_csv(
    'Data/Inputs_FertilizersNutrient_E_All_Data/Inputs_FertilizersNutrient_E_All_Data_NOFLAG.csv',
    low_memory=False
)
print(f"  {len(df_inputs):,} rows")

print("Loading bilateral trade data...")
df_trade = pd.read_csv(
    'Data/Fertilizers_DetailedTradeMatrix_E_All_Data/Fertilizers_DetailedTradeMatrix_E_All_Data_NOFLAG.csv',
    low_memory=False
)
print(f"  {len(df_trade):,} rows")

# ??????????????????????????????????????????????????????????????????????????????
# Build datasets
# ??????????????????????????????????????????????????????????????????????????????
P_pre_all, C_pre_all = {}, {}       # 2007 production & demand
P_post_all, C_post_all = {}, {}     # 2009 production & demand
T0_all = {}                          # avg 2005-2007 trade matrix
T_post_all = {}                      # 2009 actual trade

for nut, ic in NUTRIENT_CODES.items():
    P_pre_all[nut]  = load_series_single(df_inputs, ic, ELEMENT_CODES['Production'], PRE_YEAR)
    C_pre_all[nut]  = load_series_single(df_inputs, ic, ELEMENT_CODES['AgUse'],      PRE_YEAR)
    P_post_all[nut] = load_series_single(df_inputs, ic, ELEMENT_CODES['Production'], POST_YEAR)
    C_post_all[nut] = load_series_single(df_inputs, ic, ELEMENT_CODES['AgUse'],      POST_YEAR)
    T0_all[nut]     = load_trade_matrix(df_trade, ic, ELEMENT_CODES['Export'], T0_YEARS)
    T_post_all[nut] = load_trade_single(df_trade, ic, ELEMENT_CODES['Export'], POST_YEAR)

print(f"\n?? Data Summary ??")
print(f"  T? baseline: avg {T0_YEARS}")
print(f"  Pre-shock:   {PRE_YEAR}")
print(f"  Post-shock:  {POST_YEAR}\n")
print(f"{'Nutrient':20s} {'Pre prod':>10s} {'Post prod':>10s} {'T? shape':>12s} {'T_post shape':>14s}")
print("?" * 70)
for nut in ['N', 'P', 'K']:
    npre  = (P_pre_all[nut] > 1000).sum()
    npost = (P_post_all[nut] > 1000).sum()
    s0 = T0_all[nut].shape
    sp = T_post_all[nut].shape
    print(f"{NUTRIENT_NAMES[nut]:20s} {npre:>10d} {npost:>10d} "
          f"{s0[0]:>5d}?{s0[1]:<5d}  {sp[0]:>5d}?{sp[1]:<5d}")
```

## 2. The Real Shock ? Production Changes 2007 ? 2009

What actually happened to fertilizer production during the financial crisis?

```python
for nut in ['N', 'P', 'K']:
    P_pre  = P_pre_all[nut]
    P_post = P_post_all[nut]
    C_pre  = C_pre_all[nut]
    C_post = C_post_all[nut]

    print(f"?? {NUTRIENT_NAMES[nut]} ? Production: {PRE_YEAR} ? {POST_YEAR} ??\n")

    # Show ALL countries with >15% drop and >50 kt
    drops = []
    for c in sorted(set(P_pre.index) | set(P_post.index)):
        pre = P_pre.get(c, 0)
        post = P_post.get(c, 0)
        if pre > 50000:
            chg = post - pre
            pct = (chg / pre * 100) if pre > 0 else 0
            drops.append((c, pre, post, chg, pct))

    drops.sort(key=lambda x: x[3])
    
    print(f"{'Country':40s} {'2007 (t)':>12s} {'2009 (t)':>12s} {'Change (t)':>14s} {'%':>8s}")
    print("?" * 90)
    for c, pre, post, chg, pct in drops[:20]:
        marker = " ???" if pct < -15 else ""
        print(f"{c:40s} {pre:>12,.0f} {post:>12,.0f} {chg:>+14,.0f} {pct:>+7.1f}%{marker}")

    g_pre, g_post = P_pre.sum(), P_post.sum()
    print(f"\n{'GLOBAL TOTAL':40s} {g_pre:>12,.0f} {g_post:>12,.0f} "
          f"{g_post-g_pre:>+14,.0f} {(g_post/g_pre-1)*100:>+7.1f}%")

    # Demand changes for shock countries
    print(f"\n{'Country':40s} {'Demand 2007':>12s} {'Demand 2009':>12s} {'%':>8s}")
    print("?" * 65)
    for c in SHOCK_COUNTRIES:
        pre_d = C_pre.get(c, 0)
        post_d = C_post.get(c, 0)
        if pre_d > 1000:
            pct = ((post_d / pre_d - 1) * 100) if pre_d > 0 else 0
            print(f"{c:40s} {pre_d:>12,.0f} {post_d:>12,.0f} {pct:>+7.1f}%")
    print()
```

## 3. RAS Algorithm

```python
def run_ras(T0, S_hat, D_hat, max_iter=1_000, tol=1e-6, verbose=True):
    """RAS / Iterative Proportional Fitting."""
    countries = T0.index.tolist()
    T = T0.to_numpy(copy=True).astype(float)
    s = S_hat.to_numpy(copy=True)
    d = D_hat.to_numpy(copy=True)
    r = np.ones(len(countries))
    c = np.ones(len(countries))

    for iteration in range(1, max_iter + 1):
        row_sums = (r[:, None] * T * c[None, :]).sum(axis=1)
        with np.errstate(invalid='ignore', divide='ignore'):
            r_new = np.where(row_sums > 0, r * s / row_sums, r)
        col_sums = (r_new[:, None] * T * c[None, :]).sum(axis=0)
        with np.errstate(invalid='ignore', divide='ignore'):
            c_new = np.where(col_sums > 0, c * d / col_sums, c)

        X_current = r_new[:, None] * T * c_new[None, :]
        max_err = max(
            np.abs(X_current.sum(axis=1) - s).max(),
            np.abs(X_current.sum(axis=0) - d).max()
        )
        r, c = r_new, c_new
        if max_err < tol:
            if verbose:
                print(f"  RAS converged in {iteration} iters (err = {max_err:.2e})")
            break
    else:
        if verbose:
            print(f"  WARNING: RAS did not converge after {max_iter} iters (err = {max_err:.2e})")

    return pd.DataFrame(r[:, None] * T * c[None, :], index=countries, columns=countries)


def run_pipeline(P_vec, C_vec, T0_mat, country_list):
    """Phase 1 (domestic filter) + Phase 2 (normalize) + Phase 3 (RAS)."""
    nc = len(country_list)
    K = pd.Series(np.zeros(nc), index=country_list)
    S = pd.Series(np.zeros(nc), index=country_list)
    D = pd.Series(np.zeros(nc), index=country_list)
    for i in country_list:
        bi = P_vec[i] - C_vec[i]
        if bi > 0:
            K[i], S[i], D[i] = C_vec[i], bi, 0
        else:
            K[i], S[i], D[i] = P_vec[i], 0, abs(bi)
    st, dt = S.sum(), D.sum()
    S_hat, D_hat = S.copy(), D.copy()
    if st < dt:
        D_hat = D * (st / dt)
    elif st > dt:
        S_hat = S * (dt / st)
    X = run_ras(T0_mat, S_hat, D_hat, verbose=False)
    return X, K

print("RAS functions defined.")
```

## 4. Run RAS & Compare with Actual 2009 Trade

Three scenarios:
1. **Baseline**: Pre-shock P & C (2007) ? sanity check
2. **RAS predicted (post P, pre C)**: Only production changes
3. **RAS predicted (post P & C)**: Both production and demand change

Compare all three with **actual 2009 trade**.

```python
results_all = {}

for nut in ['N', 'P', 'K']:
    print(f"\n{'?'*70}")
    print(f"  {NUTRIENT_NAMES[nut]}")
    print(f"{'?'*70}")

    P_pre  = P_pre_all[nut]
    C_pre  = C_pre_all[nut]
    P_post = P_post_all[nut]
    C_post = C_post_all[nut]

    all_c = sorted(set(P_pre.index) | set(C_pre.index) | set(P_post.index) | set(C_post.index))
    P_pre  = P_pre.reindex(all_c, fill_value=0).clip(lower=0)
    C_pre  = C_pre.reindex(all_c, fill_value=0).clip(lower=0)
    P_post = P_post.reindex(all_c, fill_value=0).clip(lower=0)
    C_post = C_post.reindex(all_c, fill_value=0).clip(lower=0)

    mask = (P_pre >= MIN_THRESHOLD) | (C_pre >= MIN_THRESHOLD) | \
           (P_post >= MIN_THRESHOLD) | (C_post >= MIN_THRESHOLD)
    ctries = sorted(P_pre[mask].index)
    nc = len(ctries)

    P_pre  = P_pre.reindex(ctries, fill_value=0)
    C_pre  = C_pre.reindex(ctries, fill_value=0)
    P_post = P_post.reindex(ctries, fill_value=0)
    C_post = C_post.reindex(ctries, fill_value=0)

    # Baseline T0
    T0 = T0_all[nut].reindex(index=ctries, columns=ctries, fill_value=0)
    _v = T0.to_numpy(copy=True); np.fill_diagonal(_v, 0)
    T0 = pd.DataFrame(_v, index=T0.index, columns=T0.columns)

    # Actual 2009 trade
    Ta = T_post_all[nut].reindex(index=ctries, columns=ctries, fill_value=0)
    _v2 = Ta.to_numpy(copy=True); np.fill_diagonal(_v2, 0)
    Ta = pd.DataFrame(_v2, index=Ta.index, columns=Ta.columns)

    # Run scenarios
    print(f"  {nc} countries")
    print(f"  Scenario 1: Baseline (2007 P & C)...")
    X_base, K_base = run_pipeline(P_pre, C_pre, T0, ctries)
    print(f"  Scenario 2: Post P (2009), Pre C (2007)...")
    X_pred1, K_pred1 = run_pipeline(P_post, C_pre, T0, ctries)
    print(f"  Scenario 3: Post P & C (both 2009)...")
    X_pred2, K_pred2 = run_pipeline(P_post, C_post, T0, ctries)

    # Country-level metrics
    imp_base   = X_base.sum(axis=0)
    imp_pred   = X_pred2.sum(axis=0)
    imp_actual = Ta.sum(axis=0)
    exp_base   = X_base.sum(axis=1)
    exp_pred   = X_pred2.sum(axis=1)
    exp_actual = Ta.sum(axis=1)

    m_imp = (imp_pred > 100) & (imp_actual > 100)
    m_exp = (exp_pred > 100) & (exp_actual > 100)

    r_imp = imp_pred[m_imp].corr(imp_actual[m_imp]) if m_imp.sum() > 2 else np.nan
    r_exp = exp_pred[m_exp].corr(exp_actual[m_exp]) if m_exp.sum() > 2 else np.nan
    mae_imp = (imp_pred[m_imp] - imp_actual[m_imp]).abs().mean()
    mae_exp = (exp_pred[m_exp] - exp_actual[m_exp]).abs().mean()

    results_all[nut] = {
        'ctries': ctries, 'nc': nc,
        'X_base': X_base, 'X_pred1': X_pred1, 'X_pred2': X_pred2, 'Ta': Ta,
        'P_pre': P_pre, 'P_post': P_post, 'C_pre': C_pre, 'C_post': C_post,
        'T0': T0,
        'imp_base': imp_base, 'imp_pred': imp_pred, 'imp_actual': imp_actual,
        'exp_base': exp_base, 'exp_pred': exp_pred, 'exp_actual': exp_actual,
        'r_imp': r_imp, 'r_exp': r_exp, 'mae_imp': mae_imp, 'mae_exp': mae_exp,
    }

    print(f"\n  {'Scenario':45s} {'Total Trade (t)':>16s}")
    print(f"  {'?'*63}")
    print(f"  {'Baseline RAS (2007 P & C)':45s} {X_base.values.sum():>16,.0f}")
    print(f"  {'RAS predicted (2009 P, 2007 C)':45s} {X_pred1.values.sum():>16,.0f}")
    print(f"  {'RAS predicted (2009 P & C)':45s} {X_pred2.values.sum():>16,.0f}")
    print(f"  {'Actual 2009 trade':45s} {Ta.values.sum():>16,.0f}")
    print(f"\n  r(imports) = {r_imp:.4f}   r(exports) = {r_exp:.4f}")
    print(f"  MAE(imports) = {mae_imp:,.0f} t   MAE(exports) = {mae_exp:,.0f} t")
```

## 5b. Comprehensive Error Analysis

We evaluate the RAS model at three levels:

**1. Aggregate** ? Does RAS get the total volume right?  
**2. Country level** ? Per-country imports/exports accuracy  
**3. Bilateral** ? Individual route-level flows  

For each level we compute:
- **R?** ? proportion of variance explained (perfect = 1.0)
- **MAPE** ? Mean Absolute Percentage Error (interpretable %)
- **Direction-of-change accuracy** ? Did RAS correctly predict increase/decrease vs baseline?
- **Theil U** ? Decomposed into bias, variance, and covariance (where do errors come from?)
- **Log-scale analysis** ? Trade spans orders of magnitude; log-log fit reveals structural accuracy

```python
# ??????????????????????????????????????????????????????????????????????????????
# Error analysis functions
# ??????????????????????????????????????????????????????????????????????????????
def compute_metrics(predicted, actual, baseline=None, min_val=100):
    """Compute comprehensive error metrics between predicted and actual series."""
    mask = (predicted.abs() > min_val) | (actual.abs() > min_val)
    p = predicted[mask].fillna(0)
    a = actual[mask].fillna(0)

    n = len(p)
    if n < 3:
        return {}

    # Basic
    r_pearson = p.corr(a)
    r_spearman = p.rank().corr(a.rank())

    residuals = p - a
    mae = residuals.abs().mean()
    rmse = np.sqrt((residuals ** 2).mean())

    # R?
    ss_res = (residuals ** 2).sum()
    ss_tot = ((a - a.mean()) ** 2).sum()
    r_squared = 1 - ss_res / ss_tot if ss_tot > 0 else np.nan

    # MAPE (only where actual > threshold to avoid division by ~0)
    mape_mask = a.abs() > 1000
    if mape_mask.sum() > 0:
        mape = ((p[mape_mask] - a[mape_mask]).abs() / a[mape_mask].abs() * 100).mean()
    else:
        mape = np.nan

    # Median Absolute Percentage Error (more robust than MAPE)
    if mape_mask.sum() > 0:
        mdape = ((p[mape_mask] - a[mape_mask]).abs() / a[mape_mask].abs() * 100).median()
    else:
        mdape = np.nan

    # Theil U decomposition
    mse = (residuals ** 2).mean()
    if mse > 0:
        bias_prop = (p.mean() - a.mean()) ** 2 / mse
        var_prop = (p.std() - a.std()) ** 2 / mse
        cov_prop = 1 - bias_prop - var_prop
    else:
        bias_prop = var_prop = cov_prop = 0

    # Direction-of-change accuracy (vs baseline)
    doc = np.nan
    if baseline is not None:
        b = baseline.reindex(p.index, fill_value=0)
        change_pred = p - b
        change_actual = a - b
        both_nonzero = (change_pred.abs() > 100) | (change_actual.abs() > 100)
        if both_nonzero.sum() > 0:
            same_sign = (change_pred[both_nonzero] * change_actual[both_nonzero]) >= 0
            doc = same_sign.mean() * 100

    return {
        'n': n,
        'pearson_r': r_pearson,
        'spearman_r': r_spearman,
        'R?': r_squared,
        'MAE': mae,
        'RMSE': rmse,
        'MAPE': mape,
        'MdAPE': mdape,
        'Theil_bias%': bias_prop * 100,
        'Theil_var%': var_prop * 100,
        'Theil_cov%': cov_prop * 100,
        'Dir_accuracy%': doc,
    }

# ??????????????????????????????????????????????????????????????????????????????
# Run error analysis for all nutrients at all levels
# ??????????????????????????????????????????????????????????????????????????????
print("?? Comprehensive Error Analysis ? 2009 Financial Crisis ??\n")

for nut in ['N', 'P', 'K']:
    r = results_all[nut]
    print(f"{'?'*80}")
    print(f"  {NUTRIENT_NAMES[nut]}")
    print(f"{'?'*80}")

    # --- Country-level imports ---
    m_imp = compute_metrics(r['imp_pred'], r['imp_actual'], r['imp_base'])
    print(f"\n  Country-level IMPORTS ({m_imp['n']} countries):")
    print(f"    Pearson r:        {m_imp['pearson_r']:.4f}")
    print(f"    Spearman ?:       {m_imp['spearman_r']:.4f}")
    print(f"    R?:               {m_imp['R?']:.4f}")
    print(f"    MAPE:             {m_imp['MAPE']:.1f}%")
    print(f"    MdAPE:            {m_imp['MdAPE']:.1f}%")
    print(f"    MAE:              {m_imp['MAE']:,.0f} t")
    print(f"    RMSE:             {m_imp['RMSE']:,.0f} t")
    print(f"    Direction accur.: {m_imp['Dir_accuracy%']:.1f}%")
    print(f"    Theil decomp:     bias={m_imp['Theil_bias%']:.1f}%, "
          f"variance={m_imp['Theil_var%']:.1f}%, covariance={m_imp['Theil_cov%']:.1f}%")

    # --- Country-level exports ---
    m_exp = compute_metrics(r['exp_pred'], r['exp_actual'], r['exp_base'])
    print(f"\n  Country-level EXPORTS ({m_exp['n']} countries):")
    print(f"    Pearson r:        {m_exp['pearson_r']:.4f}")
    print(f"    Spearman ?:       {m_exp['spearman_r']:.4f}")
    print(f"    R?:               {m_exp['R?']:.4f}")
    print(f"    MAPE:             {m_exp['MAPE']:.1f}%")
    print(f"    MdAPE:            {m_exp['MdAPE']:.1f}%")
    print(f"    Direction accur.: {m_exp['Dir_accuracy%']:.1f}%")
    print(f"    Theil decomp:     bias={m_exp['Theil_bias%']:.1f}%, "
          f"variance={m_exp['Theil_var%']:.1f}%, covariance={m_exp['Theil_cov%']:.1f}%")

    # --- Bilateral flows ---
    def flatten_for_metrics(mat, label):
        rows = []
        for i in mat.index:
            for j in mat.columns:
                if i != j:
                    rows.append({'pair': f"{i}|{j}", label: mat.loc[i,j]})
        return pd.DataFrame(rows).set_index('pair')[label]

    bi_pred = flatten_for_metrics(r['X_pred2'], 'pred')
    bi_act  = flatten_for_metrics(r['Ta'], 'actual')
    bi_base = flatten_for_metrics(r['X_base'], 'base')
    m_bi = compute_metrics(bi_pred, bi_act, bi_base, min_val=500)
    print(f"\n  Bilateral FLOWS ({m_bi['n']} flows > 500 t):")
    print(f"    Pearson r:        {m_bi['pearson_r']:.4f}")
    print(f"    R?:               {m_bi['R?']:.4f}")
    print(f"    MAPE:             {m_bi['MAPE']:.1f}%")
    print(f"    MdAPE:            {m_bi['MdAPE']:.1f}%")
    print(f"    Direction accur.: {m_bi['Dir_accuracy%']:.1f}%")
    print()
```

## 6b. Fertilizer Availability per Country ? % of Pre-Shock Level

For each country we compute:
- **F_baseline** = domestic kept + imports under baseline (2007) conditions
- **F_RAS** = domestic kept + imports as predicted by RAS (post-shock production & demand)
- **F_actual** = domestic kept + actual imports in 2009

Then: `Avail_vs_baseline_%` = F / F_baseline × 100

This shows the **percentage reduction in fertilizer availability** compared to pre-shock levels.

```python
# ??????????????????????????????????????????????????????????????????????????
# Fertilizer Availability as % of baseline ? RAS vs Actual
# ??????????????????????????????????????????????????????????????????????????

for nut in ['N', 'P', 'K']:
    r = results_all[nut]
    ctries = r['ctries']
    P_pre, P_post = r['P_pre'], r['P_post']
    C_pre, C_post = r['C_pre'], r['C_post']

    # Baseline availability: K_base + imports_base
    K_base = pd.Series(0.0, index=ctries)
    for c in ctries:
        K_base[c] = min(P_pre[c], C_pre[c])
    imp_base = r['X_base'].sum(axis=0)
    F_base = K_base + imp_base

    # RAS-predicted availability: K_ras + imports_ras
    K_ras = pd.Series(0.0, index=ctries)
    for c in ctries:
        K_ras[c] = min(P_post[c], C_post[c])
    imp_ras = r['X_pred2'].sum(axis=0)
    F_ras = K_ras + imp_ras

    # Actual 2009 availability: K_actual + imports_actual
    K_actual = K_ras.copy()  # same post-shock production ? same domestic kept
    imp_actual = r['Ta'].sum(axis=0)
    F_actual = K_actual + imp_actual

    # Build comparison table
    avail = pd.DataFrame({
        'Demand': C_post,
        'F_baseline': F_base,
        'F_RAS': F_ras,
        'F_actual': F_actual,
        'RAS_vs_base_%': (F_ras / F_base * 100).round(1),
        'Actual_vs_base_%': (F_actual / F_base * 100).round(1),
        'RAS_reduction_%': ((1 - F_ras / F_base) * 100).round(1),
        'Actual_reduction_%': ((1 - F_actual / F_base) * 100).round(1),
        'Error_pp': ((F_ras / F_base - F_actual / F_base) * 100).round(1),
    }).replace([np.inf, -np.inf], np.nan).dropna()

    # Only countries with meaningful baseline
    avail = avail[avail['F_baseline'] > 1000].copy()
    avail = avail.sort_values('Actual_reduction_%', ascending=False)

    print(f"\n{'?'*90}")
    print(f"  {NUTRIENT_NAMES[nut]} ? Availability as % of Baseline")
    print(f"{'?'*90}")

    print(f"\n  Top 15 countries with LARGEST availability reduction (actual):")
    display(avail.head(15)[['F_baseline', 'F_RAS', 'F_actual',
                            'RAS_vs_base_%', 'Actual_vs_base_%',
                            'RAS_reduction_%', 'Actual_reduction_%', 'Error_pp']].round(0))

    # Scatter: RAS-predicted % vs Actual %
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    ax = axes[0]
    x = avail['Actual_vs_base_%'].clip(0, 200)
    y = avail['RAS_vs_base_%'].clip(0, 200)
    ax.scatter(x, y, alpha=0.5, s=20, c='#2980b9')
    lims = [0, max(x.max(), y.max()) * 1.05]
    ax.plot(lims, lims, 'r--', alpha=0.5, label='Perfect prediction')
    ax.set_xlabel('Actual availability (% of baseline)')
    ax.set_ylabel('RAS predicted availability (% of baseline)')
    ax.set_title(f'{NUTRIENT_NAMES[nut]} ? RAS vs Actual')
    ax.legend()

    # Error distribution
    ax2 = axes[1]
    errors = avail['Error_pp'].dropna()
    ax2.hist(errors.clip(-50, 50), bins=40, color='#e67e22', alpha=0.7, edgecolor='white')
    ax2.axvline(0, color='red', linestyle='--', alpha=0.7)
    ax2.set_xlabel('Error (RAS % ? Actual %, pp)')
    ax2.set_ylabel('Number of countries')
    ax2.set_title(f'Prediction error distribution\nMedian: {errors.median():.1f} pp, MAE: {errors.abs().mean():.1f} pp')

    plt.suptitle(f'{NUTRIENT_NAMES[nut]} ? Fertilizer Availability: RAS Prediction vs Reality',
                 fontsize=13, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    # Summary stats
    r_corr = avail['RAS_vs_base_%'].corr(avail['Actual_vs_base_%'])
    mae = avail['Error_pp'].abs().mean()
    median_err = avail['Error_pp'].median()
    print(f"\n  Correlation (RAS% vs Actual%): r = {r_corr:.3f}")
    print(f"  Mean Absolute Error: {mae:.1f} pp")
    print(f"  Median Error: {median_err:+.1f} pp (+ = RAS overschat beschikbaarheid)")
```

## 8b. Genormaliseerde Veranderingsscore per Route

Per route schalen we de verandering zodat:
- **0** = baseline (T?, avg 2005?2007) ? geen verandering t.o.v. pre-shock
- **1** = actual 2009 ? de werkelijke post-shock realiteit

$$\text{score} = \frac{\text{RAS} - \text{Baseline}}{\text{Actual}_{2009} - \text{Baseline}}$$

| Score | Betekenis |
|-------|-----------|
| **1.0** | RAS voorspelt de werkelijke verandering perfect |
| **0.0** | RAS voorspelt g??n verandering (blijft op baseline) |
| **0.5** | RAS vangt de helft van de werkelijke verschuiving op |
| **>1** | RAS overschiet (verandert m??r dan werkelijk) |
| **<0** | RAS gaat de verkeerde richting op |

```python
nut = 'N'
r = results_all[nut]
X_pred = r['X_pred2']
X_base = r['X_base']
T_act  = r['Ta']
T0     = r['T0']

# ??????????????????????????????????????????????????????????????????????????????
# Build edge table with the 0-1 normalized score
# ??????????????????????????????????????????????????????????????????????????????
edges = []
for i in X_pred.index:
    for j in X_pred.columns:
        if i == j:
            continue
        base = T0.loc[i, j]
        ras  = X_pred.loc[i, j]
        act  = T_act.loc[i, j]
        if act > 500 or ras > 500 or base > 500:
            delta = act - base
            if abs(delta) > 50:
                score = (ras - base) / delta
            else:
                score = np.nan  # no meaningful change to measure
            edges.append({
                'from': i, 'to': j,
                'baseline': base,
                'RAS': ras,
                'actual': act,
                'delta': delta,
                'score': score,
            })

edf = pd.DataFrame(edges)
edf['in_baseline'] = edf['baseline'] > 100

# Only keep routes with a meaningful change (|delta| > 500 t)
edf_sig = edf[edf['delta'].abs() > 500].copy().dropna(subset=['score'])
edf_sig = edf_sig.sort_values('score')

n_total = len(edf_sig)
print(f"?? Veranderingsscore per Route ? {NUTRIENT_NAMES[nut]} ??")
print(f"   {n_total} routes met significante verandering (|delta| > 500 t)\n")
print(f"   Score: 0 = baseline (geen verandering), 1 = actual 2009 (perfect)\n")

# Summary
scores = edf_sig['score'].clip(-5, 5)
print(f"   Mediaan score:  {scores.median():.2f}")
print(f"   Gemiddeld:      {scores.mean():.2f}")
print(f"   % score 0.7?1.3  (goed):       {((scores >= 0.7) & (scores <= 1.3)).mean()*100:.0f}%")
print(f"   % score 0.3?1.7  (redelijk):   {((scores >= 0.3) & (scores <= 1.7)).mean()*100:.0f}%")
print(f"   % score < 0       (verkeerde richting): {(scores < 0).mean()*100:.0f}%")
print(f"   % score > 2       (overshoot):  {(scores > 2).mean()*100:.0f}%")

# ??????????????????????????????????????????????????????????????????????????????
# Full table: baseline=0, actual=1, RAS=score
# ??????????????????????????????????????????????????????????????????????????????
# Sort by absolute deviation from perfect (1.0)
edf_sig['abs_err'] = (edf_sig['score'] - 1).abs()

# Show routes with biggest changes first (most material)
edf_big = edf_sig.nlargest(50, 'delta' if edf_sig['delta'].max() > 0 else 'abs_err')
edf_big = edf_big.sort_values('delta')  # biggest declines first

print(f"\n?? TABEL: Top 50 routes op verandering ? 0=baseline, 1=actual, RAS=score ??\n")
print(f"{'From':25s} {'To':25s} ?{'Base (kt)':>10s} {'Act (kt)':>10s} {'RAS (kt)':>10s} ?"
      f"{'? (kt)':>9s} ? {'0':>4s}{'???score???':^13s}{'1':>4s} ? {'Score':>6s}")
print("?" * 125)

for _, row in edf_big.iterrows():
    sc = row['score']
    # Visual bar: 0----score----1
    bar_len = 20
    if np.isnan(sc):
        bar = " " * bar_len
        sc_str = "  n/a"
    else:
        sc_clipped = max(-1, min(3, sc))
        # Map -1..3 to 0..bar_len
        pos = int((sc_clipped + 1) / 4 * bar_len)
        zero_pos = int(1 / 4 * bar_len)   # position of 0
        one_pos = int(2 / 4 * bar_len)    # position of 1
        bar_chars = list('?' * bar_len)
        bar_chars[zero_pos] = '0'
        bar_chars[one_pos] = '1'
        if 0 <= pos < bar_len:
            bar_chars[pos] = '?'
        bar = ''.join(bar_chars)
        sc_str = f"{sc:>+6.2f}"

    print(f"{row['from'][:24]:25s} {row['to'][:24]:25s} ?"
          f"{row['baseline']/1e3:>10.1f} {row['actual']/1e3:>10.1f} {row['RAS']/1e3:>10.1f} ?"
          f"{row['delta']/1e3:>+8.1f} ? {bar} ? {sc_str}")
```

## 8c. Uitschieters Verklaren

De grootste uitschieters zeggen ons waar het proportionele model faalt. We categoriseren ze:

- **Sanction/Policy**: route veranderde door beleid, niet markt
- **New entrant**: nieuwe producent of nieuwe handelsrelatie
- **Demand shift**: importerend land schakelde naar andere leverancier
- **Capacity**: productiecapaciteit veranderde niet-proportioneel

```python
# ??????????????????????????????????????????????????????????????????????????????
# Uitschieters met productie/vraag context
# ??????????????????????????????????????????????????????????????????????????????
P_pre  = r['P_pre']
P_post = r['P_post']
C_pre  = r['C_pre']
C_post = r['C_post']

def add_context(df):
    rows = []
    for _, row in df.iterrows():
        fr, to = row['from'], row['to']
        p_pre_v  = P_pre.get(fr, 0)
        p_post_v = P_post.get(fr, 0)
        c_pre_v  = C_pre.get(to, 0)
        c_post_v = C_post.get(to, 0)
        p_chg = ((p_post_v / p_pre_v - 1) * 100) if p_pre_v > 0 else 0
        c_chg = ((c_post_v / c_pre_v - 1) * 100) if c_pre_v > 0 else 0
        rows.append({**row.to_dict(), 'exp_prod_chg%': p_chg, 'imp_dem_chg%': c_chg})
    return pd.DataFrame(rows)

# Worst outliers: farthest from score=1
edf_sig['abs_score_err'] = (edf_sig['score'] - 1).abs()
worst = edf_sig.nlargest(30, 'abs_score_err')
worst_ctx = add_context(worst).sort_values('score')

print(f"?? 30 Grootste Uitschieters (verst van score=1) ? {NUTRIENT_NAMES[nut]} ??\n")
print(f"{'From':22s} {'To':22s} ?{'Base':>8s} {'RAS':>8s} {'Act':>8s} ?"
      f" {'Score':>6s} ? {'ExpProd%':>9s} {'ImpDem%':>9s} ? Interpretatie")
print("?" * 130)

for _, row in worst_ctx.iterrows():
    sc = row['score']
    # Auto-interpretatie
    if sc < 0:
        interp = "? verkeerde richting"
    elif sc < 0.3:
        interp = "? nauwelijks verandering voorspeld"
    elif 0.3 <= sc < 0.7:
        interp = "~ onderschat verandering"
    elif 0.7 <= sc <= 1.3:
        interp = "? goed"
    elif 1.3 < sc <= 2:
        interp = "~ overschat verandering"
    else:
        interp = "? sterk overschat"

    print(f"{row['from'][:21]:22s} {row['to'][:21]:22s} ?"
          f"{row['baseline']/1e3:>7.0f}k {row['RAS']/1e3:>7.0f}k {row['actual']/1e3:>7.0f}k ?"
          f" {sc:>+6.2f} ? {row['exp_prod_chg%']:>+8.1f}% {row['imp_dem_chg%']:>+8.1f}% ? {interp}")

print(f"""
Leeswijzer:
  Score    = (RAS ? Baseline) / (Actual ? Baseline)
             0 = RAS ziet geen verandering, 1 = perfect, <0 = verkeerde richting
  ExpProd% = productieverandering exporteur (2007?2009)
  ImpDem%  = vraagverandering importeur (2007?2009)

Patronen:
  ? Score ? 1 + groot ExpProd% ? RAS vangt productieschok goed op
  ? Score ? 0 + route is nieuw  ? RAS kan geen nieuwe routes maken
  ? Score < 0                   ? structurele verschuiving die RAS niet kan modelleren""")
```

## 9b. Landbalans: Productie ? Export + Import ? Agrarisch Gebruik

Per land per jaar berekenen we de **nutri?ntbalans**:

$$\text{Balans} = \underbrace{P}_{\text{productie}} - \underbrace{E}_{\text{export}} + \underbrace{I}_{\text{import}} - \underbrace{A}_{\text{agrarisch gebruik}}$$

- **Balans > 0**: het land houdt nutri?nt over (voorraadopbouw, industrieel gebruik, verlies)
- **Balans < 0**: het land verbruikt meer dan het ontvangt (voorraadafbouw)
- **Balans ? 0**: alles wordt gebruikt of verhandeld

Dit toont het "normale" patroon en maakt afwijkingen rond 2008-2009 zichtbaar.

```python
# ??????????????????????????????????????????????????????????????????????????????
# Build complete balance table: P - E + I - A per country per year per nutrient
# ??????????????????????????????????????????????????????????????????????????????
BALANCE_YEARS = list(range(2002, 2024))
bal_ycols = [f'Y{y}' for y in BALANCE_YEARS]

for col in bal_ycols:
    if col in df_inputs.columns:
        df_inputs[col] = pd.to_numeric(df_inputs[col], errors='coerce')

def get_yearly(df, item_code, elem_code):
    mask = (df['Item Code'] == item_code) & (df['Element Code'] == elem_code)
    sub = df.loc[mask, ['Area'] + bal_ycols].copy()
    return sub.groupby('Area')[bal_ycols].sum()

balance_all = {}
for nut, ic in NUTRIENT_CODES.items():
    prod = get_yearly(df_inputs, ic, 5510)   # Production
    exp  = get_yearly(df_inputs, ic, 5910)   # Export
    imp  = get_yearly(df_inputs, ic, 5610)   # Import
    aguse = get_yearly(df_inputs, ic, 5157)  # Agricultural Use

    # Align all to same country set
    all_countries = sorted(set(prod.index) | set(exp.index) | set(imp.index) | set(aguse.index))
    prod  = prod.reindex(all_countries, fill_value=0)
    exp   = exp.reindex(all_countries, fill_value=0)
    imp   = imp.reindex(all_countries, fill_value=0)
    aguse = aguse.reindex(all_countries, fill_value=0)

    bal = prod - exp + imp - aguse

    balance_all[nut] = {
        'prod': prod, 'exp': exp, 'imp': imp, 'aguse': aguse, 'balance': bal
    }

print("Balance data built for all nutrients and years (2002-2023).")
print(f"Countries: {len(all_countries)}")
```

## 11. Interpretation

### Why 2009 is a better test for nuclear war modeling

| Property | 2009 Crisis | 2022 Russia-Ukraine | Nuclear war |
|----------|-------------|---------------------|-------------|
| **Shock type** | Production shutdown | Sanctions / trade barriers | Production destruction |
| **What RAS models** | ? Production change | ? Trade barriers | ? Production change |
| **Multiple countries** | ? Russia, Belarus, Israel, Germany... | ? Mainly Belarus exports | ? Multiple belligerents |
| **New trade routes needed?** | Less so (demand fell too) | Yes (sanctions rerouting) | Yes (but supply scarce) |
| **Demand effect** | Demand also dropped | Demand stable | Demand collapses (nuclear winter) |

### Implications for catastrophic risk modeling

- If RAS performs **well** for 2009: the model is suitable for production-shock scenarios (including nuclear war), where the fundamental assumption of proportional trade adjustment holds
- If RAS performs **poorly** for 2009: even genuine production shocks trigger non-proportional trade restructuring, and the model needs augmentation regardless of shock type
- Compare the **convergence behavior**: does RAS converge for 2009 (unlike for 2022)?
